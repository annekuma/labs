````python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
````

````python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
 print(y_test[i],'\n')
 print(X_test[i],'\n')
 plt.contourf(np.rot90(X_test[i].transpose()))
 plt.show()
````

````python
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

````python
import pandas as pd
import numpy as np

Z_train = []
for i in range(60000):
    row = X_train[i].reshape((1, 784))
    Z_train.append(row[0])
Z_train = pd.DataFrame(Z_train)
Z_test = []
for i in range(len(y_test)):
    row = X_test[i].reshape((1, 784))
    Z_test.append(row[0])
Z_test = pd.DataFrame(Z_test)
````

````python
from sklearn.neighbors import KNeighborsClassifier

k_bar = 50
accuracy = np.zeros(k_bar)

for k in range(1, k_bar + 1):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(Z_train.values, y_train)
    accuracy[k - 1] = knn.score(Z_test.values, y_test)

k_star = np.argmax(accuracy) + 1
accuracy_max = accuracy[k_star - 1]

print(f"Optimal k: {k_star} with accuracy: {accuracy_max:.4f}")
plt.plot(range(1, k_bar + 1), accuracy)  # Adjust x-axis to reflect k values starting from 1
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title(f"Optimal k: {k_star}")
plt.show()
````

````python
knn = KNeighborsClassifier(n_neighbors=1)
predictor = knn.fit(Z_train.values,y_train)
y_hat = predictor.predict(Z_test.values)
accuracy = knn.score(Z_test.values,y_test)
print('Accuracy: ', accuracy)
pd.crosstab(y_test, y_hat)
````
